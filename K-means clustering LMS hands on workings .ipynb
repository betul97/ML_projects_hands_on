```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv("iris (2).csv")
iris.head()
```

```python
iris.species.unique()
```

```python
sns.scatterplot(data=iris,x="sepal_length",y="petal_width",hue="species",palette="coolwarm_r")
```

```python
from sklearn.cluster import KMeans
```

```python
X=iris[["sepal_length","petal_width"]]
km=KMeans(n_clusters=3,n_init=3,init="random",random_state=42)
km.fit(X)
y_kmeans= km.predict(X)
y_kmeans
```

```python
import sklearn
sklearn.__version__
```

```python
sns.scatterplot(data=iris,x="sepal_length",y="petal_width",hue=y_kmeans,palette="coolwarm_r")
centers=km.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c="black",s=200,alpha=0.6)
```

```python
km.inertia_
```

```python
new_data=[[4.7,0.8]]
y_pred=km.predict(new_data)
y_pred
```

```python
sns.scatterplot(data=iris,x="sepal_length",y="petal_width")
```

```python
intertia=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(X)
    intertia.append(km.inertia_)
    
plt.plot(K,intertia,marker="x")
plt.xlabel("k")
plt.xticks(np.arange(15))
plt.ylabel("intertia")
plt.title("Elbow Method")
plt.show
```
